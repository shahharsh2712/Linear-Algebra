2. Shortest distance

```r
y <- matrix(c(-1,-5,10), nrow=3)
y
```

```r
u1 <- matrix(c(5,-2,1), nrow=3)
u1
```

```r
u2 <- matrix(c(1,2,-1), nrow=3)
u2
```

```r
a <- dot(y,u1)
b <- dot(u1,u1)
c <- dot(y,u2)
d <- dot(u2,u2)
```

```r
y_bar <- (a/b)*u1 + (c/d)*u2
y_bar
```

```r
dist <- norm(y - y_bar, '2')
round(dist, 2)
```

 Shortest distance is 6.71

3. Mileage

```r
data <- read.csv('mileage.csv')
head(data)
```

```r
disp <- sqrt(data$DISP)
disp <- as.vector(disp)
```

```r
disp <- 1/disp
disp
```

```r
x <- as.matrix(cbind(rep(1, nrow(data)), disp))
head(x)
```

```r
y <- as.matrix(data$MILEAGE)
head(y)
```

```r
const = inv(t(x) %*% x) %*% t(x) %*% y
const
```

```r
cat("Model parameters are ", const)
```

```r
lsq_error1 <- norm(y - x %*% const, '2')
cat("The least square error in Model is ", lsq_error1)
```

```r
test_disp <- 1.8
```

```r
model_pred <- c(1, test_disp) %*% const
model_pred
```

```r
cat("Using Model, mpg of car is", model_pred)
```
